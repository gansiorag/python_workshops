```python
def next_level(kk: list):
    str_w = ''    
    if len(kk)>0:
        for kk1 in kk:
            print(kk1.tag)
            f_rez = open('rez_parser.txt', 'a')
            f_rez.write(kk1.tag + '\n')
            f_rez.close()
            if len(kk1.attrib)>0:
                f_rez = open('rez_parser.txt', 'a')
                for dd in kk1.attrib:
                    print('atr => ', dd, ' => ', kk1.attrib[dd])
                    f_rez.write('atr => ' + dd + ' => ' + kk1.attrib[dd] + '\n')
                f_rez.close()
            if kk1.text: 
                #srt_w = kk1.tag + '  atr=> ' + kk1.attrib + '  text=> ' + kk1.text + '\n'
                #print(type(kk1.tag))
                print('  text=> ',kk1.text)
                f_rez = open('rez_parser.txt', 'a')
                f_rez.write('text => ' + kk1.text +'\n')
                f_rez.close()
            next_level(kk1)
```
