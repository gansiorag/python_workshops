```python
def playlist_item(url):
    headers = {
        'authority': 'api.youtubemultidownloader.com',
        'sec-ch-ua': '" Not A;Brand";v="99", "Chromium";v="98", "Yandex";v="22"',
        'accept': 'application/json, text/javascript, */*; q=0.01',
        'dnt': '1',
        'sec-ch-ua-mobile': '?0',
        'user-agent': random.choice(random.choice(list_user_agents)),
        'sec-ch-ua-platform': '"Windows"',
        'origin': 'https://youtubemultidownloader.net',
        'sec-fetch-site': 'cross-site',
        'sec-fetch-mode': 'cors',
        'sec-fetch-dest': 'empty',
        'referer': 'https://youtubemultidownloader.net/',
        'accept-language': 'ru,en;q=0.9,uk;q=0.8',
    }

    params = {
        'url': url,
        'nextPageToken': '',
    }

    response = requests.get('https://api.youtubemultidownloader.com/playlist', params=params, headers=headers).json()
    list_items = []
    for item in range(0, len(response['items'])):
        list_items.append(response['items'][item]['id'])
    return list_items
```

```python
def get_all_playlist(id_channel):
    headers = {
        'authority': 'api.youtubemultidownloader.com',
        'sec-ch-ua': '" Not A;Brand";v="99", "Chromium";v="98", "Yandex";v="22"',
        'accept': 'application/json, text/javascript, */*; q=0.01',
        'dnt': '1',
        'sec-ch-ua-mobile': '?0',
        'user-agent': random.choice(random.choice(list_user_agents)),
        'sec-ch-ua-platform': '"Windows"',
        'origin': 'https://youtubemultidownloader.net',
        'sec-fetch-site': 'cross-site',
        'sec-fetch-mode': 'cors',
        'sec-fetch-dest': 'empty',
        'referer': 'https://youtubemultidownloader.net/',
        'accept-language': 'ru,en;q=0.9,uk;q=0.8',
    }

    params = {
        #'url': url,
        'nextPageToken': '',
    }
    # Print the full request URL and parameters for debugging
    print(f"Request URL: https://api.youtubemultidownloader.com/{id_channel} with params: {params}")

    try:
        response = requests.get(f'https://api.youtubemultidownloader.com/{id_channel}', params=params, headers=headers)
        response.raise_for_status()  # Raise an error for bad responses
        data = response.json()

        # Check for the expected response structure
        if 'items' in data:
            list_items = [item['id'] for item in data['items']]
            return list_items
        
        # Handle the case where the API response does not contain 'items'
        print(f"Unexpected response structure: {data}")
        return None

    except requests.exceptions.RequestException as e:
        print(f"An error occurred: {e}")
        return None
```

```python
def get_base_info_video(videoId, playlistId):
    # Базовая информация по видео
    base_video_url = 'https://www.googleapis.com/youtube/v3/videos'
    headers = {'accept': 'application/json'}
    params = {

        'part': 'snippet,contentDetails,statistics',
        'id': videoId,
        'key': USER_KEY
    }
    answ_video = requests.get(url=base_video_url, params=params, headers = headers)
    all_data = answ_video.json()
    base_data = {
             'name_channel': all_data['items'][0]['snippet']['channelTitle'],
             'channel_id': all_data['items'][0]['snippet']['channelId'],        
            'videoId': videoId,
             'playlistId': playlistId,
             'name_file': all_data['items'][0]['snippet']['title'],
             'description': all_data['items'][0]['snippet']['description'].replace('\u2063', ''),
             'publishedAt': all_data['items'][0]['snippet']['publishedAt'],
             'duration': all_data['items'][0]['contentDetails']['duration'],
             'viewCount': all_data['items'][0]['statistics']['viewCount'],
             'likeCount': all_data['items'][0]['statistics']['likeCount'],
             'favoriteCount': all_data['items'][0]['statistics']['favoriteCount'],
             'commentCount': all_data['items'][0]['statistics']['commentCount'],
              }
    return base_data
```

```python
def get_all_comments(base_data, name_folder, id_video_file):
    # Получение всех комментариев к видео
    k = 0
    url = 'https://www.googleapis.com/youtube/v3/commentThreads'
    headers = {'Accept': 'application/json'}
    params = {
        'key': USER_KEY,
        'textFormat': 'plainText',
        'part': 'snippet,replies',
        'videoId': base_data['videoId'],
        'maxResults': 100
    }
    answ_inw = requests.get(url=url, headers = headers, params=params).json()
    print(answ_inw)
    with open(name_folder + '/' + f'comment_{str(k)}_{id_video_file}.json', 'w') as f:
        json.dump(answ_inw, f)
    # Получение следующей страницы комментариев
    while 'nextPageToken' in answ_inw:
        k += 1
        print(k)
        url_next_page = 'https://www.googleapis.com/youtube/v3/commentThreads'
        params_next = {
            'key': USER_KEY,
            'textFormat': 'plainText',
            'part': 'snippet,replies',
            'videoId': base_data['videoId'],
            'maxResults': 100,
            'pageToken':answ_inw['nextPageToken']
        }
        answ_inw = requests.get(url=url_next_page, headers = headers, params=params_next).json()
        print(answ_inw)
        with open(name_folder + '/' + f'comment_{str(k)}_{id_video_file}.json', 'w') as f:
            json.dump(answ_inw, f)
        


    
```
