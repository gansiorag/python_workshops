```python
from typing import Any, Optional, List, Dict
from pydantic import BaseModel, ConfigDict 
```

```python
# Правильное объявление типа словаря: ключ - строка, значение - целое число
Dd = Dict[str, int] 

# Пример использования
my_dict: Dd = {"apple": 5, "banana": 10}
print(my_dict)
```

```python
class ConnectBase(BaseModel):
    type_base: str
    connect: Any = None  # Задать значения по умолчанию, если нужно
    cursor: Any = None

    # Конструктор, который принимает аргументы
    def __init__(self, type_base: str, **data):
        super().__init__(type_base=type_base, **data)
        
        # Логика выполняется только при создании объекта
        if self.type_base == 'Sqlite':
            self.connect = 'Sqlite'
            print(self.connect)
            
        elif self.type_base == 'PosgreSQL': 
            self.cursor = 'PosgreSQL'
            print(self.cursor) # Исправлено с ccursor на cursor
```

```python
# --- Пример использования ---
# Создаем объект для SQLite
sqlite_conn = ConnectBase(type_base='Sqlite')
# Вывод: Sqlite

# Создаем объект для PostgreSQL
pg_conn = ConnectBase(type_base='PosgreSQL')
# Вывод: PosgreSQL
```
