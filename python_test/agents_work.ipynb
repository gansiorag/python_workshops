```python
%autoawait
# import asyncio
from pydantic_ai import Agent
from pydantic_ai.providers.ollama import OllamaProvider # Ollama Provider get url
from pydantic_ai.models.openai import OpenAIChatModel #  OpenAI Provider get model name
from pydantic_ai.capabilities import Thinking, WebSearch # Here’s the same agent with thinking and web search capabilities:
```

```python
from pydantic_ai import providers

print(dir(providers))
```

```python

ollama_provider = OllamaProvider(base_url='http://localhost:11434/v1')
ollama_model = OpenAIChatModel(
    model_name='llama3.1:8b',
    provider=ollama_provider
)

```

```python
instruction = """Ты программист на python экспертного уровня. Отвечая на вопосы
всегда приводишь не менее 5 примеров кода и не менее 5 ссылок на подтверждение своих
выводов."""


```

```python
agent = Agent(ollama_model, 
              instructions=instruction,
              capabilities=[Thinking()]
              )
```

```python
result = await agent.run('Привет, что ты умеешь?')
print(result.output)
```

```python
result = await agent.run('Расскажи о библиотеке pydantic')
print(result.output)
```
